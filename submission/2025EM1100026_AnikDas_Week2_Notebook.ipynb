## 1. Problem Statement

**Title:** Employee Attrition Prediction & Workforce Risk Analysis

Employee attrition — the voluntary or involuntary departure of employees from an organisation — is one of the most costly challenges faced by the retail industry. High staff turnover in retail disrupts operational continuity, increases recruitment and training costs, and leads to declining customer satisfaction due to loss of experienced personnel.

This project applies machine learning to a retail HR dataset to:
1. **Predict** which employees are most likely to leave the organisation.
2. **Identify** the key drivers behind attrition (e.g., salary, department, commute distance, job role).
3. **Provide** data-driven insights and retention recommendations for store management.

### Business Goals
- Identify the key drivers of voluntary employee turnover in a retail context.
- Build and compare multiple classification models to predict employee attrition risk.
- Understand attrition reasons and geographic/departmental patterns across the workforce.
- Provide actionable recommendations to store management to improve staff retention.

## 5. Initial Observations & Data Quality Analysis

### Initial Analysis of Inspection Output:
- **Data Volume:** The dataset contains **9,000 employee records across 31 features**, providing a rich and comprehensive sample for reliable model training and evaluation. The increased feature count gives significantly more dimensions for feature engineering and EDA compared to standard attrition datasets.
- **Data Types:** Features are a mix of numeric (integer), categorical (string), ordinal, and date types. `Date_of_Joining` is loaded as `str` and requires conversion to `datetime` for feature extraction. `Environment_Satisfaction` is text-based ordinal (Very High / High / Medium / Low) and requires ordinal encoding.
- **Missing Values:** Only two columns contain missing values, and both are **intentionally missing by design**:
  - `Attrition_Reason`: 8,345 missing (92.7%) — NaN only for employees who have **not** left (`Attrition_Status = 'No'`). This is expected and logical.
  - `Attrition_Date`: 8,345 missing (92.7%) — NaN only for employees who have **not** left. Same reason as above.
  - **No imputation is needed.** These columns will be dropped before model training to prevent target leakage.
  - All 29 remaining predictor columns have **zero missing values**.
- **Class Imbalance:** The target variable `Attrition_Status` shows significant imbalance — **92.72% No (8,345) vs 7.28% Yes (655)**, a ratio of approximately 12.7:1. This is realistic for retail sector attrition. Standard accuracy will be a highly misleading metric; **F1-Score (for class 'Yes') and ROC-AUC are the primary metrics**. Class imbalance will be handled via SMOTE applied only to the training set to prevent data leakage.
- **Rich New Features:** This dataset includes geolocation (`State`, `City`, `Distance_from_Home`), employment context (`Employment_Type`, `Work_Location`, `Position`), detailed performance history (`Last_3_Year_Rating`, `Performance_Score`), cost data (`Employee_Cost_to_Company`, `Annual_Income`), and most importantly **`Attrition_Reason`** — providing 10 specific reasons for departure, enabling root-cause analysis beyond standard binary classification.

## 6. Proposed Preprocessing & Feature Engineering Workflow

### Step 1 — Data Cleaning & Column Handling:
- Drop `Employee_ID` and `Employee_Name` (non-predictive identifiers — names are anonymised as 'Employee_N').
- Drop `Attrition_Reason` and `Attrition_Date` **before model training** to prevent target leakage (these columns only have values when `Attrition_Status = 'Yes'`).
- Parse `Date_of_Joining` to datetime; extract `Joining_Year` and `Joining_Month` as engineered features.
- Verify data integrity: no missing values exist in any predictor column — **no imputation required**.

### Step 2 — Exploratory Data Analysis (EDA):
- Univariate analysis: distributions of `Age`, `Monthly_Income`, `Years_at_Company`, `Distance_from_Home`, `Yearly_Total_Training_Hours`.
- Bivariate analysis: `Attrition_Status` vs `Employment_Type`, `Department`, `Job_Role`, `Work_Location`, `Position`, `Distance_from_Home`.
- Attrition root-cause analysis using `Attrition_Reason` (for the 655 attrited employees).
- Geographic attrition analysis: attrition rate by `State` and `Work_Location`.
- Correlation heatmap of all numeric features.
- `Attrition_Status` rate by `Department`, `State`, and `Work_Location` (grouped bar charts and heatmaps).

### Step 3 — Feature Engineering:
- `IncomeToRoleRatio`: `Monthly_Income` / median income of the same `Job_Role`.
- `SalaryPerYear`: `Monthly_Income × 12` / `Employee_Cost_to_Company` — captures salary-to-CTC efficiency ratio as a retention risk signal.
- `TenureBand`: bin `Years_at_Company` into groups (1–3, 4–6, 7–9) aligned with the dataset range of 1–9 years.
- `CommuteBand`: bin `Distance_from_Home` into Short (< 15 km), Medium (15–30 km), Long (> 30 km).
- `Joining_Year` and `Joining_Month` extracted from `Date_of_Joining`.

### Step 4 — Encoding & Scaling:
- **One-Hot Encode** nominals: `Gender`, `Department`, `Job_Role`, `Employment_Type`, `Work_Location`, `Education_Field`, `Marital_Status`, `State`, `Onboarding_Status`, `Last_3_Year_Rating`.
- **Ordinal Encode**: `Education_Level` (High School < Diploma < Graduate < Postgraduate < PhD), `Position` (Junior < Mid-Level < Senior < Lead), `Performance_Score` (B < A < A+ < A*), `Environment_Satisfaction` (Low < Medium < High < Very High).
- **StandardScaler** on all numeric features.

### Step 5 — Class Imbalance Handling:
- Apply **SMOTE** only on training data using `imblearn.pipeline` (prevents data leakage).
- Alternative: `class_weight='balanced'` for tree-based models.

### Step 6 — Model Building & Evaluation:
- Train/test split: 80% train, 20% test (stratified by `Attrition_Status`).
- Train all 5 proposed models and record metrics.
- Evaluate using F1-Score (Yes class), ROC-AUC, Precision, Recall, Confusion Matrix.

## 7. Proposed Modelling & Evaluation Plan

### Proposed Pipeline:
- **Baseline Model:** Logistic Regression — high interpretability, good linear decision boundary.
- **Candidate Advanced Models:**
  - Decision Tree Classifier — captures non-linear relationships; feature importance
  - Random Forest Classifier — ensemble of trees; robust to overfitting
  - Gradient Boosting Classifier — typically best performer on tabular classification
  - K-Nearest Neighbors (KNN) — distance-based comparison model

### Evaluation Strategy:
- Primary metrics: **F1-Score (for Class 'Yes')** and **ROC-AUC** — robust under class imbalance.
- Supporting metrics: Precision, Recall, Confusion Matrix.
- Hyperparameter optimisation: GridSearchCV (Logistic Regression, Decision Tree) and RandomizedSearchCV (Random Forest, Gradient Boosting).
- Cross-validation: 5-fold Stratified K-Fold throughout.
