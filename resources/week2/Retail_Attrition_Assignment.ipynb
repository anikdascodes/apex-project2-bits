## 5. Initial Observations & Data Quality Analysis

### Initial Analysis of Inspection Output:
- **Data Volume:** The dataset contains 10,000 observations, providing a sufficiently large sample size for reliable model training and validation.
- **Data Types:** Features are a mix of numerical, categorical, and ordinal types. The `Date_of_Joining` variable is currently loaded as an object and will require conversion to a datetime format during preprocessing.

## 6. Proposed Preprocessing & Feature Engineering Workflow

### 1. Data Cleaning and Imputation:
- Impute missing value in `MonthlyIncome` using the median value grouped by `JobRole`.
- Impute missing ratings in satisfaction metrics (`JobSatisfaction`, `EnvironmentSatisfaction`, `WorkLifeBalance`) using the respective median or mode.
- Handle missing categorical entries in `Onboarding_Status` using the most frequent value (mode).

### 2. Feature Engineering:
- Extract features from `Date_of_Joining` (such as joining year and month) to analyze tenure trends.
- Compute relative pay metrics by comparing individual incomes to the median income of their specific `JobRole`.
- Create career progression metrics such as promotion-to-tenure ratios.

### 3. Categorical Encoding:
- Use One-Hot Encoding for nominal variables (`Gender`, `Department`, `JobRole`, `Onboarding_Status`).
- Apply Label/Ordinal encoding for ordinal variables such as `Education`.

### 4. Feature Scaling:
- Standardize numeric features using StandardScaler to ensure distance-based models perform correctly.
